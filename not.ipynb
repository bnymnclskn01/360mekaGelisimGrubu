```python
def harfnotu(ortalama):
    if ortalama >= 90 and ortalama <= 100:
        hnotu = "AA"
    elif ortalama >= 85 and ortalama < 90:
        hnotu = "BA"
    elif ortalama >= 80 and ortalama < 85:
        hnotu = "BB"
    elif ortalama >= 75 and ortalama < 80:
        hnotu = "CB"
    elif ortalama >= 70 and ortalama < 75:
        hnotu = "CC"
    elif ortalama >= 60 and ortalama < 70:
        hnotu = "DC"
    elif ortalama >= 50 and ortalama < 60:
        hnotu = "DD"
    else:
        hnotu = "F"
    return hnotu
```

```python
import time
def zaman(ogrenciler):
    tsl = time.localtime()
    anlik = time.strftime("%c",tsl)
    gun = anlik[8:10]
    ay = anlik[4:7]
    yil = anlik[20:24]
    saat = anlik[11:13]
    dakika = anlik[14:16]
    return gun, ay, yil, saat, dakika
```

```python
ogrenciler = {}
cikis = False
while not cikis:
    adsoyad = input("Öğrencinin Adını ve Soyadını giriniz: ")
    if adsoyad == "FFFF":
        cikis = True
        break
    elif adsoyad not in ogrenciler:
        vize = [0,0,0,0,0,0]
        final = [0,0,0,0,0,0]
        but = [0,0,0,0,0,0]
        hnotu = " "
        ogrenciler[adsoyad] = [vize, final, but, hnotu]
        tamam1 = False
    else:
        print("Öğrencinin notu zaten girilmiştir.")
    
    
    while not tamam1:
        vizenotu = input("Öğrencinin Vize notunu giriniz: ")
        if vizenotu == "":
            tamam1 = True
            tamam2 = False
        elif vizenotu.isdigit() and int(vizenotu) <= 100 and int(vizenotu) >=0:
            tamam1 = True
            zbilgi = zaman(ogrenciler)
            vize = [int(vizenotu),zbilgi[0],zbilgi[1],zbilgi[2],zbilgi[3],zbilgi[4]]
            eskideger = ogrenciler[adsoyad]
            eskideger[0] = vize
            tamam2 = False
        else:
            print("Hatalı bir değer girdiniz lütfen 0-100 aralığında bir değer giriniz")
    
    
    while not tamam2:
        finalnotu = input("Öğrencinin Final notunu giriniz: ")
        if finalnotu == "":
            tamam2 = True
            tamam3 = False
        elif finalnotu.isdigit() and int(finalnotu) <= 100 and int(finalnotu) >=0:
            tamam2 = True
            tamam3 = False
            zbilgi = zaman(ogrenciler)
            final = [int(finalnotu),zbilgi[0],zbilgi[1],zbilgi[2],zbilgi[3],zbilgi[4]]
            eskideger = ogrenciler[adsoyad]
            eskideger[1] = final
        else:
            print("Hatalı bir değer girdiniz lütfen 0-100 aralığında bir değer giriniz")
    
    
    while not tamam3:
        butnotu = input("Öğrencinin Bütünleme notunu giriniz: ")
        if butnotu == "":
            tamam3 = True
        elif butnotu.isdigit() and int(butnotu) <= 100 and int(butnotu) >=0:
            tamam3 = True
            zbilgi = zaman(ogrenciler)
            but = [int(butnotu),zbilgi[0],zbilgi[1],zbilgi[2],zbilgi[3],zbilgi[4]]
            eskideger = ogrenciler[adsoyad]
            eskideger[2] = but
        else:
            print("Hatalı bir değer girdiniz lütfen 0-100 aralığında bir değer giriniz")
            
    vizenotu = vize[0]
    finalnotu = final[0]
    butnotu = but[0]
    
    if butnotu <= finalnotu:
        ort = 0.4*vizenotu + 0.6*finalnotu
    else:
        ort = 0.4*vizenotu + 0.6*butnotu
    harfnot = harfnotu(ort)
    eskideger = ogrenciler[adsoyad]
    eskideger[3] = harfnot
print(ogrenciler)
```
